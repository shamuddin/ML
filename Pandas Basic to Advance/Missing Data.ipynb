**Reindex() creats a copy(not a view)**

```python
browser_index = ['Firefox','Chrome','Safari','IELO','Koqueror']
browser_df = pd.DataFrame(
    {
        'http_status':[200,200,404,404,301],
        'Response Time':[0.04,0.02,0.07,0.08,1.0],
    },
    index=browser_index
)
browser_df
```

**Creating a new index**

```python
new_index = ['Safari','Seahorse','ComodoDragon','IELO','Chrome']
broswer_df2 = browser_df.reindex(new_index)
broswer_df2
```

**Drop all the missing values**

```python
broswer_df3 = broswer_df2.dropna(how='any')
broswer_df3
```

**Fill in missing data**

```python
broswer_df2.fillna(value=0.05)
```
