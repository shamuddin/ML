```python
import pandas as pd
import numpy as np
```

```python
starting_date = '20200301'
sample_numpy_data = np.array(np.arange(50)).reshape(10,5)
sample_numpy_data
```

```python
dates_index = pd.date_range(starting_date,periods=10)
sample_df = pd.DataFrame(sample_numpy_data,index=dates_index,columns=list('ABCDE'))
sample_df
```

```python
second_numpy_data = np.array(np.arange(len(sample_df))) *100 + 7
sample_df['F']=second_numpy_data
sample_df
```
